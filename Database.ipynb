```python
import sqlite3
import csv

# Connect to the SQLite database (creates a new database if it doesn't exist)
conn = sqlite3.connect('database1.db')
cursor = conn.cursor()
print ("Opened database successfully")

# Create the 'songs' table if it doesn't exist
cursor.execute('''
    CREATE TABLE IF NOT EXISTS songs (
        Title TEXT,
        Artist TEXT,
        URL TEXT
    )
''')

# Read data from the CSV file
csv_file_path = 'us_songs.csv'  # Replace with your actual CSV file path

with open(csv_file_path, 'r') as file:
    csv_reader = csv.reader(file)
    
    # Skip the header row if it exists
    next(csv_reader, None)

    # Insert data into the 'songs' table
    cursor.executemany('INSERT INTO songs VALUES (?, ?, ?)', csv_reader)
print ("Table created successfully")
# Commit the changes and close the connection
conn.commit()
conn.close()
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('database1.db')
cursor = conn.cursor()

# Select all rows from the 'songs' table
cursor.execute('SELECT * FROM songs')
rows = cursor.fetchall()

# Print the data
for row in rows:
    print(row)

# Close the connection
conn.close()
```

```python
import sqlite3
import csv

# Connect to the SQLite database (creates a new database if it doesn't exist)
conn = sqlite3.connect('database1.db')
cursor = conn.cursor()
print("Opened database successfully")

# Create the 'global_songs' table if it doesn't exist
cursor.execute('''
    CREATE TABLE IF NOT EXISTS global_songs (
        Title TEXT,
        Artist TEXT,
        URL TEXT
    )
''')

# Read data from the CSV file
csv_file_path = 'global_songs.csv'  # Replace with your actual CSV file path

with open(csv_file_path, 'r') as file:
    csv_reader = csv.reader(file)

    # Skip the header row if it exists
    next(csv_reader, None)

    # Insert data into the 'global_songs' table
    cursor.executemany('INSERT INTO global_songs VALUES (?, ?, ?)', csv_reader)

print("Table created successfully")

# Commit the changes and close the connection
conn.commit()
conn.close()
```
